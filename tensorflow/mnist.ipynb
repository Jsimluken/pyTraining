```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
import os
#os.environ['KMP_DUPLICATE_LIB_OK']='TRUE'
```

```python
mnist = input_data.read_data_sets("mnist_data", one_hot=True)
```

```python
x = tf.placeholder(tf.float32, [None, 784])
y = tf.placeholder(tf.float32, [None, 10])
h = tf.layers.dense(x, 128, activation=tf.nn.sigmoid)
h = tf.layers.dense(h, 64, activation=tf.nn.sigmoid)
output = tf.layers.dense(h, 10, activation=tf.nn.softmax)
```

```python
loss = tf.reduce_mean(-tf.reduce_sum(y*tf.log(output), axis=1))
print(loss)
optimizer = tf.train.AdamOptimizer().minimize(loss)
```

```python
epochs = 10000
with tf.Session() as sess:
    tf.global_variables_initializer().run()
    for i in range (epochs):
        batch_xs, batch_ys = mnist.train.next_batch(128)
        sess.run(optimizer, feed_dict={x:batch_xs, y:batch_ys})
    batch_xa, batch_ya = mnist.test.next_batch(100)
    pred = sess.run(output, feed_dict={x:batch_xa, y:batch_ya})
```
